In this competition, we need to identify to which of the three classes (EAP, HPL, or MWS) each sentence belongs to. To solve this text classification problem, Kaggle has specified multi-class log-loss as evaluation metric. This is implemented in the [following way](https://github.com/dnouri/nolearn/blob/master/nolearn/lasagne/util.py):

```python
def multiclass_logloss(actual, predicted, eps=1e-15):
    """Multi class version of Logarithmic Loss metric.
    :param actual: Array containing the actual target classes
    :param predicted: Matrix with class predictions, one probability per class
    """
    #Convert 'actual' to a binary array if it is not already
    if len(actual.shape) == 1:
        actual2 = np.zeros((actual.shape[0], predicted.shape[1]))
        for i, val in enumerate(actual):
            actual2[i, val] = 1
        actual = actual2

    clip = np.clip(predicted, eps, 1 - eps)
    rows = actual.shape[0]
    vsota = np.sum(actual * np.log(clip))
    return -1.0 / rows * vsota
```

# Data preparation

We will go through these pre-processing steps to convert the input raw text into a usable format:

### Tokenization
Segregation of the text into its individual constitutent words.
### Stopwords
Throw away any words that occur too frequently as its frequency of occurrence will not be useful in helping detecting relevant texts (we can also consider throwing away words that occur very infrequently).
### Stemming
combine variants of words into a single parent word that still conveys the same meaning
### Vectorization
Converting text into vector format. One of the simplest is the famous bag-of-words approach, where you create a matrix (for each document or text in the corpus). In the simplest form, this matrix stores word frequencies (word counts) and is oft referred to as vectorization of the raw text.

The LabelEncoder from scikit-learn is used to convert text labels to integers 0, 1, or 2

```python
lable_encoder = preprocessing.LabelEncoder()
y = lable_encoder.fit_transform(train.author.values)
```

Next, we split the data into training and validation sets

```python
xtrain, xvalid, ytrain, yvalid = train_test_split(
    train.text.values,
    y,
    stratify=y,
    random_state=12,
    test_size=0.2,
    shuffle=True)
```

# Natural Language Processing
## Building a base model
Our first model is a TF-IDF (Term Frequency - Inverse Document Frequency) followed by a simple Logistic Regression.

```python
tfv = TfidfVectorizer(
    min_df=3,
    max_features=None, 
    strip_accents='unicode',
    analyzer='word',
    token_pattern=r'\w{1,}',
    ngram_range=(1,3),
    use_idf=1,
    smooth_idf=1,
    sublinear_tf=1,
    stop_words='english'
)

# Fitting TF-IDF to both training and test sets (semi-supervised learning)
tfv.fit(list(xtrain) + list(xvalid))
xtrain_tfv =  tfv.transform(xtrain) 
xvalid_tfv = tfv.transform(xvalid)
```

```python
# Fitting a Logistic Regression on TF-IDF
clf = LogisticRegression(C=1.0, max_iter=200)
clf.fit(xtrain_tfv, ytrain)
predictions = clf.predict_proba(xvalid_tfv)

print ("logloss: %0.3f " % multiclass_logloss(yvalid, predictions))
```

Now let's try to improve the logloss we obtained from this simple model. Replacing word counts with TF-IDF will improve the logloss value from 0.583 to 0.534:

```python
ctv = CountVectorizer(
    analyzer='word',
    token_pattern=r'\w{1,}',
    ngram_range=(1,3),
    stop_words = 'english'
)

# Fitting Count Vectorizer to both training and validation sets
ctv.fit(list(xtrain) + list(xvalid))
xtrain_ctv =  ctv.transform(xtrain) 
xvalid_ctv = ctv.transform(xvalid)
```

```python
# Fitting a simple Logistic Regression on Counts
clf = LogisticRegression(C=1.0)
clf.fit(xtrain_ctv, ytrain)
predictions = clf.predict_proba(xvalid_ctv)

print ("logloss: %0.3f " % multiclass_logloss(yvalid, predictions))
```

# Hyperparameter optimization
A hyperparameter optimization could help with the results. Here I use hyperopt for Tree of Parzen Estimators (TPE) optimization. Let's try it on the xgboost method with tf-idf svd features as an example:

```python
space_xgboost = {
    'xgboost_max_depth' : hp.choice('xgboost_max_depth', [5, 7, 9]),
    'xgboost_n_estimators' : hp.choice('xgboost_n_estimators', [150, 180, 210]),
    'xgboost_colsample_bytree': hp.uniform('xgboost_colsample_bytree', 0.5, 0.9),
    'xgboost_subsample': hp.uniform('xgboost_subsample', 0.5, 0.9),
    'xgboost_learning_rate': hp.uniform('xgboost_learning_rate', 0.01, 0.2),
}
```

```python
def create_xgboost_pipeline(params):
    clf = xgb.XGBClassifier(
        max_depth=params['xgboost_max_depth'],
        n_estimators=params['xgboost_n_estimators'],
        colsample_bytree=params['xgboost_colsample_bytree'],
        subsample=params['xgboost_subsample'],
        nthread=10,
        use_label_encoder=False,
        learning_rate=params['xgboost_learning_rate'],
        verbosity=0
    )
    clf.fit(xtrain_svd, ytrain)
    predictions = clf.predict_proba(xvalid_svd)
    logloss = multiclass_logloss(yvalid, predictions)
    return {'loss': logloss, 'status': STATUS_OK}
```

```python
trials = Trials()
best_params = fmin(create_xgboost_pipeline, space_xgboost, algo=tpe.suggest, max_evals=10, trials=trials)
print ('best run: ', best_params)
```

The best obtained logloss out of the 10 tried cases was 0.74. This for sure can be improved further by modifying the optimization parameters.

# Word vectors
Now let's create sentence vectors and use them to create a machine learning model. We can try GloVe vectors, word2vec, or fasttext. In this post, I'll be using the word2vec vectors. Word2Vec is an algorithm that accepts text corpus as an input and outputs a vector representation for each word. Given a set of sentences (corpus) the model loops on the words of each sentence and either tries to use the current word w in order to predict its neighbors (i.e., its context), this approach is called “Skip-Gram”, or it uses each of these contexts to predict the current word w, in that case the method is called “Continuous Bag Of Words” (CBOW).

Gensim comes with several already pre-trained models, let's try glove-twitter-25

```python
print(list(gensim.downloader.info()['models'].keys()))
```

```python
#Download the "word2vec-google-news-300" embeddings
w2v_vectors = gensim.downloader.load('word2vec-google-news-300')
```

```python
#This function creates a normalized vector for the whole sentence
def sent2vec(s, num_vec):
    words = str(s).lower()
    words = word_tokenize(words)
    words = [w for w in words if not w in stop_words]
    words = [w for w in words if w.isalpha()]
    M = []
    for w in words:
        try:
            M.append(w2v_vectors.get_vector(w))
        except:
            continue
    M = np.array(M)
    v = M.sum(axis=0)
    if type(v) != np.ndarray:
        return np.zeros(num_vec)
    return v/np.sqrt((v**2).sum())
```

```python
#Convert the input data to vectors
xtrain_w2v = [sent2vec(x, 300) for x in xtrain]
xvalid_w2v = [sent2vec(x, 300) for x in xvalid]
```

```python
xtrain_w2v = np.array(xtrain_w2v)
xvalid_w2v = np.array(xvalid_w2v)
```

## Hyperparameter optimization
The results can be imroved by some hyperparameter optimization.

```python
space_dnn = {
    'num_layers': hp.choice('num_layers', ['four', 'six', 'eight']),
    'units1': hp.uniform('units1', 128, 512),
    'units2': hp.uniform('units2', 128, 512),
    'units3': hp.uniform('units3', 128, 512),
    'units4': hp.uniform('units4', 128, 512),
    'units6': hp.uniform('units6', 128, 512),
    'units8': hp.uniform('units8', 128, 512),
    'dropout': hp.uniform('dropout', 0.1, 0.4),
    'batch_size' : hp.choice('batch_size', [32, 64, 128]),
    'learning_rate': hp.uniform('learning_rate', 0.0002, 0.005),
}
```

```python
def create_dnn_pipeline(params):
    model = Sequential()
    
    model.add(Dense(units=int(params['units1']), input_dim = 300))
    model.add(Activation('relu'))
    model.add(BatchNormalization())
    model.add(Dropout(params['dropout']))

    model.add(Dense(units=int(params['units2'])))
    model.add(Activation('relu'))
    model.add(BatchNormalization())
    model.add(Dropout(params['dropout']))
    
    model.add(Dense(units=int(params['units3'])))
    model.add(Activation('relu'))
    model.add(BatchNormalization())
    model.add(Dropout(params['dropout']))
    
    model.add(Dense(units=int(params['units4'])))
    model.add(Activation('relu'))
    model.add(BatchNormalization())
    model.add(Dropout(params['dropout']))
    
    if params['num_layers'] != 'four':
        for i in range(2):
            model.add(Dense(units=int(params['units6'])))
            model.add(Activation('relu'))
            model.add(BatchNormalization())
            model.add(Dropout(params['dropout']))
        
    if params['num_layers'] == 'eight':
        for i in range(2):
            model.add(Dense(units=int(params['units8'])))
            model.add(Activation('relu'))
            model.add(BatchNormalization())
            model.add(Dropout(params['dropout']))

    model.add(Dense(3))
    model.add(Activation('softmax'))
    
    learning_rate = params['learning_rate']
    optimizer = optimizers.Adam(learning_rate=learning_rate)
    model.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=['mse', 'mae', 'mape'])
    
    early_stop = EarlyStopping(monitor='val_loss', min_delta=0.001, patience=nb_epochs/4, verbose=0)
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=nb_epochs/4, min_lr=0.0625*learning_rate)
    callbacks = [early_stop, reduce_lr]
    
    result = model.fit(
        xtrain_w2v_scl, 
        y=ytrain_enc,
        epochs=nb_epochs,
        verbose=0, 
        batch_size=int(params['batch_size']),
        callbacks = callbacks,
        validation_data = (xvalid_w2v_scl, yvalid_enc)
    )

    return {'loss': result.history['val_loss'][-1], 'status': STATUS_OK}
```

```python
trials_dnn = Trials()
best_params = fmin(create_dnn_pipeline, space_dnn, algo=tpe.suggest, max_evals=10, trials=trials_dnn)
print ('best run: ', best_params)
```

## LSTM

For LSTMs, we need to tokenize the text data

```python
#Using keras tokenizer here
token = text.Tokenizer(num_words=None)
max_len = 70

token.fit_on_texts(list(xtrain) + list(xvalid))
xtrain_seq = token.texts_to_sequences(xtrain)
xvalid_seq = token.texts_to_sequences(xvalid)

# zero pad the sequences
xtrain_pad = sequence.pad_sequences(xtrain_seq, maxlen=max_len)
xvalid_pad = sequence.pad_sequences(xvalid_seq, maxlen=max_len)

word_index = token.word_index
```

```python
#Create an embedding matrix for the words we have in the dataset
#For simplicity, we ignore the words that are not part of our word2vec
embedding_matrix = np.zeros((len(word_index) + 1, 300))
for word, i in word_index.items():
    if word not in stop_words and word.isalpha():
        if word in w2v_vectors.key_to_index:
            embedding_vector = w2v_vectors.get_vector(word)
            if embedding_vector is not None:
                embedding_matrix[i] = embedding_vector
```

A large dropout value helps to avoid overfitting

```python
# A simple LSTM with word2vec embeddings and two dense layers
model = Sequential()

model.add(
    Embedding(
        len(word_index) + 1,
        300,
        weights=[embedding_matrix],
        input_length=max_len,
        trainable=False
    )
)

model.add(SpatialDropout1D(0.3))
model.add(LSTM(300, dropout=0.3, recurrent_dropout=0.3))

model.add(Dense(512, activation='relu'))
model.add(Dropout(0.8))

model.add(Dense(512, activation='relu'))
model.add(Dropout(0.8))

model.add(Dense(3))
model.add(Activation('softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam')

# Fit the model with early stopping callback
earlystop = EarlyStopping(monitor='val_loss', min_delta=0, patience=3, verbose=0, mode='auto')
result = model.fit(
    xtrain_pad,
    y=ytrain_enc,
    batch_size=512,
    epochs=100, 
    verbose=1,
    validation_data=(xvalid_pad, yvalid_enc),
    callbacks=[earlystop]
)
```

```python
print(result.history['val_loss'][-1])
```

Let's see if two layers of GRU can improve the results

# Ensembling
Making ensemble of different methods can help achieve a lower loss. Let's use a very simple VotingClassifier module in sklearn as follows

```python
model1 = LogisticRegression(C=1.0, max_iter=200)
model2 = tree.DecisionTreeClassifier(random_state=1)
clf = VotingClassifier(estimators=[('lr', model1), ('dt', model2)], voting='soft')
clf.fit(xtrain_ctv, ytrain)
predictions = clf.predict_proba(xvalid_ctv)
print ("logloss: %0.3f " % multiclass_logloss(yvalid, predictions))
```
